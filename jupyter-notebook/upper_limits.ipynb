```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
%matplotlib inline
```

###   Calculations made while reading the paper by [Barlow](https://arxiv.org/abs/1905.12362) - Upper limits

- Method 4: CL$_s$

```python
r  = 3 # observed number
Nb = 3.40 # expected background events
Ns = 0 # signal events

def CLsb(Ns, Nb=3.4):
    return (poisson.pmf(0, Ns+Nb)+poisson.pmf(1, Ns+Nb)+poisson.pmf(2, Ns+Nb)+poisson.pmf(3, Ns+Nb))
```

```python
a    = np.linspace(0, 10, 41) # signal + backgorund events
blue = np.asarray([CLsb(Ns, Nb=0) for Ns in a])
red  = np.asarray([CLsb(0) for Ns in a])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.plot(a, blue*100, color="b", lw=2.5, label="CL(s+b)")
ax.plot(a, red*100, color="red", lw=2.5)

pos = np.where(a>3.4)
ax.plot(a[pos], blue[pos]/red[pos]*100, color="g", lw=2.5)
ax.axvline(3.4, color="k", ls="--")

ax.legend()
ax.set_ylabel("CL")
ax.set_xlabel("a")
```
